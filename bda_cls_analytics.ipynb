```python
#Sort DataFrame based on Date column
df = df.sort_values(by = 'Date')
```

```python
df =  df.reset_index(drop='True')
```

```python
# removing the strings in date column and leaving only time.
df['Date'] = df['Date'].apply(lambda line : re.sub('17 Aug 2020,', '', line))
df['Date'] = df['Date'].apply(lambda line : str(line).replace('GMT+5:30',''))
```

```python
df['Participant Identifier'] = df['Participant Identifier'].apply(lambda line : str(line).replace('@gvpce.ac.in',''))
```

```python
df = df.rename(columns={'Duration' : 'Duration_mins'})
```

```python

df['Duration_mins'] = df['Duration_mins'].apply(lambda line : format(line/60,'.2f'))
df['Duration_mins'] = df['Duration_mins'].apply(lambda line : float(line))
```
